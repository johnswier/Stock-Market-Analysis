```python
import pandas as pd
import csv 
```

```python
df = pd.read_csv("Multiple_year_stock_data_2014.csv")
# df =df.drop(["Unnamed: 7"], axis=1)
df.head()
```

```python
vol_list = []
ticker_list = []
yearly_change_list = []
for ticker in df["<ticker>"].unique():
    stock_vol = df.loc[(df["<ticker>"] == ticker), "<vol>"].values.sum()
    vol_list.append(stock_vol)
    
    first_entry = df.loc[(df["<ticker>"] == ticker), "<ticker>"].index[0]
    last_entry = df.loc[(df["<ticker>"] == ticker), "<ticker>"].index[-1]
    open_price = df["<open>"].iloc[first_entry]
    close_price = df["<close>"].iloc[last_entry]
    yearly_change = close_price - open_price
    yearly_change_list.append(yearly_change)
    
    ticker_list.append(ticker)
```

```python
analysis_df = pd.DataFrame({"Ticker": ticker_list, "Yearly Change": yearly_change_list, "Total Stock Volume": vol_list})
analysis_df.head(10)
```
